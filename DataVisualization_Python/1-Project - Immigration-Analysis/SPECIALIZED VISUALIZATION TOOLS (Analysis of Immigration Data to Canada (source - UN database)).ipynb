# SUBPLOTS
Often times we might want to plot multiple plots within the same figure. For example, we might want to perform a side by side comparison of the box plot with the line plot of China and India's immigration.

To visualize multiple plots together, we can create a figure (overall canvas) and divide it into subplots, each containing a plot. With subplots, we usually work with the artist layer instead of the scripting layer.

Typical syntax is :

    fig = plt.figure() # create figure
    ax = fig.add_subplot(nrows, ncols, plot_number) # create subplots



```python
fig = plt.figure() # create figure

ax0 = fig.add_subplot(1, 2, 1) # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(1, 2, 2) # add subplot 2 (1 row, 2 columns, second plot). 

# Subplot 1: Box plot
ukr_bel.plot(kind='box', color='blue', vert=False, figsize=(20, 6), ax=ax0) # add to subplot 1
ax0.set_title('Box Plots of Immigrants from Ukraine and Belarus to Canada (1993-2013)')
ax0.set_xlabel('Number of Immigrants')
ax0.set_ylabel('Countries')

# Subplot 2: Line plot
ukr_bel.plot(kind='line', figsize=(20, 6), ax=ax1) # add to subplot 2
ax1.set_title ('Line Plots of Immigrants from Ukraine and Belarus to Canada (1993 - 2013)')
ax1.set_ylabel('Number of Immigrants')
ax1.set_xlabel('Years')

plt.show()
```

# SCATTER PLOTS
(Displays values usually on two values, one is a dependent and the other is independent variable )

```python
#Use the sum() method to get the total population per year
can_total = pd.DataFrame(canad[years].sum(axis=0))
```

```python
# change the years to type int (this will be useful for regression later)
can_total.index = map(int, can_total.index)
```

```python
# Need to reset the index to put back in as a column in the can_total dataframe
can_total.reset_index(inplace = True)
```

```python
# Need to rename columns
can_total.columns = ['year', 'total']
```

```python
can_total.head()
```

```python
can_total.plot(kind='scatter', x='year', y='total', color='blue', figsize=(12,6) )
plt.title('Total Number of Immigrants to Canada (1993-2013)')
plt.xlabel('Year')
plt.ylabel('Population')

plt.show()


# Notice how the scatter plot does not connect the data points together. We can clearly observe an upward trend in the data: as the years go by, the total number of immigrants increases. We can mathematically analyze this upward trend using a regression line (line of best fit).
# So let's try to plot a linear line of best fit, and use it to predict the number of immigrants in 2016.
# Step 1: Get the equation of line of best fit. We will use Numpy's polyfit() method by passing in the following:
```

```python
x = can_total['year']      # year on x-axis
y = can_total['total']     # total on y-axis
fit = np.polyfit(x, y, deg=1)

fit
```

```python
# Step 2: Need to plot the regression line on the scatter plot.
can_total.plot(kind='scatter', x='year', y='total', figsize=(12, 6), color='blue')

plt.title('Total Immigration to Canada (1993-2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

# plot line of best fit
plt.plot(x, fit[0] * x + fit[1], color='red') # recall that x is the Years
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2000, 150000))

plt.show()

# print out the line of best fit
'No. Immigrants = {0:.0f} * Year + {1:.0f}'.format(fit[0], fit[1]) 






```

```python
#EXAMPLE
# No. Immigrants = 5567 * Year - 10926195
# No. Immigrants = 5567 * 2015 - 10926195
# No. Immigrants = 291,310

'No. Immigrants = 3910 * Year + -7606351'
'No. Immigrants = 3910 * 2016 + -7606351'
'No. Immigrants = 276,209  (Predicted Number of Future Immigratants to Canada in 2016)

```

# Bubble Plots
(A bubble plot is a variation of the scatter plot that displays three dimensions of data (x, y, z). 
 The data points are replaced with bubbles, and the size of the bubble is determined by the third variable z, 
 also known as the weight.

```python
# Let's start by analyzing the effect of Argentina's great depression.

# Argentina suffered a great depression from 1998 to 2002, which caused widespread unemployment, 
# riots, the fall of the government, and a default on the country's foreign debt. 
# In terms of income, over 50% of Argentines were poor, and seven out of ten Argentine children were 
# poor at the depth of the crisis in 2002.

# Let's analyze the effect of this crisis, and compare Argentina's immigration to that of it's neighbour Brazil. 
# Let's do that using a bubble plot of immigration from Brazil and Argentina for the years 1980 - 2013. 
# We will set the weights for the bubble as the normalized value of the population for each year.
```

```python
# Step 1: Get the data for Brazil and Argentina. 
# Like in the previous example, we will convert the Years to type int and include it in the dataframe.

#1.Get data for Argentina and Brazil
canad.loc[['Argentina', 'Brazil'], years].transpose()
```

```python
#2. Get the DataFrame
ar_br = canad.loc[['Argentina', 'Brazil'], years].transpose()
ar_br.head()
```

```python
#3. Change the Years (the index) to type int
ar_br.index = map(int, ar_br.index)
```

```python
#4. Label the index. This will automatically be the column name when we reset the index
ar_br.index.name = 'Year'
```

```python
#5. Reset index to bring the Year in as a column
ar_br.reset_index(inplace=True)
```

```python
#6. Dispaly result
ar_br.head()
```

```python
# Step 2: Create the normalized weights.

# There are several methods of normalizations in statistics, each with its own use. 
# In this case, we will use feature scaling to bring all values into the range [0, 1]. The general formula is:

# normalize Brazil data
norm_brazil = (ar_br['Brazil'] - ar_br['Brazil'].min()) / (ar_br['Brazil'].max() - ar_br['Brazil'].min())

# normalize Argentina data
norm_argentina = (ar_br['Argentina'] - ar_br['Argentina'].min()) / (ar_br['Argentina'].max() - ar_br['Argentina'].min())
```

```python
# Step 3: Plot the data.

#     To plot two different scatter plots in one plot, we can include the axes one plot into the other 
#     by passing it via the ax parameter.
#     We will also pass in the weights using the s parameter. Given that the normalized weights are between 0-1, 
# they won't be visible on the plot. Therefore, we will:
#         multiply weights by 2000 to scale it up on the graph, and,
#         add 10 to compensate for the min value (which has a 0 weight and therefore scale with ×2000

# ).

# Brazil
ax0 = ar_br.plot(kind='scatter',
                    x='Year',
                    y='Brazil',
                    figsize=(14, 8),
                    alpha=0.5,  # transparency
                    color='green',
                    s=norm_brazil * 2000 + 10,  # pass in weights 
                    xlim=(1975, 2015)
                    )

# Argentina
ax1 = ar_br.plot(kind='scatter',
                    x='Year',
                    y='Argentina',
                    alpha=0.5,
                    color="blue",
                    s=norm_argentina * 2000 + 10,
                    ax=ax0
                    )

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from Brazil and Argentina to Canada (1993-2013)')
ax0.legend(['Brazil', 'Argentina'], loc='upper left', fontsize='x-large')


# The size of the bubble corresponds to the magnitude of immigrating population for that year, 
# compared to the 1993 - 2013 data. 
# The larger the bubble is, the more immigrants are in that year.
```
