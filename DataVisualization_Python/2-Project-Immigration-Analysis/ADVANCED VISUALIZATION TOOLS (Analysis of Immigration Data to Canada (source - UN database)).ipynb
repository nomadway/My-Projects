# WAFFLE CHARTS
  (Is a way to visualize data in relation to a whole or to highlight progress against a given threshhold)

```python
canad.loc[['Kazakhstan', 'Kyrgyzstan', 'Uzbekistan'], :]
central_asia = canad.loc[['Kazakhstan', 'Kyrgyzstan', 'Uzbekistan'], :]
central_asia
```

```python
# Step 1 into creating a waffle chart is determing the proportion of each category 
# with respect to the total

# compute the proportion of each category with respect to the total
total_values = central_asia['Total'].sum()
category_proportions = central_asia['Total'] / total_values

# print out proportions
pd.DataFrame({"Category Proportion": category_proportions})
```

```python
# Step 2. Define the overall size of the waffle chart

width = 40  # width of chart
height = 10  # height of chart

total_number_tiles = width * height # total number of tiles

print(f'Total number of tiles is: {total_number_tiles}.')
```

```python
# Step 3. Use the proportion of each category to determe it respective number of tiles

# compute the number of tiles for each category
tiles_per_category = (category_proportions * total_number_tiles).round().astype(int)

# print out number of tiles per category
pd.DataFrame({"Number of tiles": tiles_per_category})
```

```python
# Step 4. Creating a matrix resembling the waffle chart and populating it

# start the waffle chart as an empty matrix
waffle_chart = np.zeros((height, width), dtype = np.uint)

# define indices to loop through waffle chart
category_index = 0
tile_index = 0

# populate the waffle chart
for col in range(width):
    for row in range(height):
        tile_index += 1

        # if the number of tiles populated for the current category is equal to its corresponding allocated tiles...
        if tile_index > sum(tiles_per_category[0:category_index]):
            # ...proceed to the next category
            category_index += 1       
            
        # set the class value to an integer, which increases with class
        waffle_chart[row, col] = category_index
        
print ('Waffle chart is populated with data from Central Asia!')
```

```python
waffle_chart
```

```python
# Step 5. Map the waffle chart matrix into a visual 

# instantiate a new figure object
new_fig = plt.figure()

# now use matshow to display the waffle chart
colormap = plt.cm.coolwarm
plt.matshow(waffle_chart, cmap=colormap)
plt.colorbar()
plt.show()

```

```python
#Step 6. Prettify the Waffle Chart

# instantiate a new figure object
new_fig = plt.figure()

# use matshow to display the waffle chart
colormap = plt.cm.coolwarm
plt.matshow(waffle_chart, cmap=colormap)
plt.colorbar()

# get the axis
ax = plt.gca()

# set minor ticks
ax.set_xticks(np.arange(-.5, (width), 1), minor=True)
ax.set_yticks(np.arange(-.5, (height), 1), minor=True)
    
# add gridlines based on minor ticks
ax.grid(which='minor', color='w', linestyle='-', linewidth=2)

plt.xticks([])
plt.yticks([])
plt.show()


```

```python
# Step 7. Create a legend and add it to chart.

# instantiate a new figure object
new_fig = plt.figure()

# use matshow to display the waffle chart
colormap = plt.cm.coolwarm
plt.matshow(waffle_chart, cmap=colormap)
plt.colorbar()

# get the axis
ax = plt.gca()

# set minor ticks
ax.set_xticks(np.arange(-.5, (width), 1), minor=True)
ax.set_yticks(np.arange(-.5, (height), 1), minor=True)
    
# add gridlines based on minor ticks
ax.grid(which='minor', color='w', linestyle='-', linewidth=2)

plt.xticks([])
plt.yticks([])

# compute cumulative sum of individual categories to match color schemes between chart and legend
values_cumsum = np.cumsum(central_asia['Total'])
total_values = values_cumsum[len(values_cumsum) - 1]
# create legend
legend_handles = []
for i, category in enumerate(central_asia.index.values):
    label_str = category + ' (' + str(central_asia['Total'][i]) + ')'
    color_val = colormap(float(values_cumsum[i])/total_values)
    legend_handles.append(mpatches.Patch(color=color_val, label=label_str))

# add legend to chart
plt.legend(handles=legend_handles,
           loc='lower center', 
           ncol=len(central_asia.index.values),
           bbox_to_anchor=(0., -0.2, 0.95, .1)
          )
plt.show()

```

```python
# Now it would very inefficient to repeat these seven steps every time we wish to create a waffle chart. 
# So let's combine all seven steps into one function called create_waffle_chart.
# This function would take the following parameters as input:

#         1.categories: Unique categories or classes in dataframe.
#         2.values: Values corresponding to categories or classes.
#         3.height: Defined height of waffle chart.
#         4.width: Defined width of waffle chart.
#         5.colormap: Colormap class
#         6.value_sign: In order to make our function more generalizable, 
#         we will add this parameter to address signs that could be associated 
#         with a value such as %, $, and so on. value_sign has a default value of empty string.
```

```python
  def create_waffle_chart(categories, values, height, width, colormap, value_sign=''):

    # compute the proportion of each category with respect to the total
    total_values = sum(values)
    category_proportions = [(float(value) / total_values) for value in values]

    # compute the total number of tiles
    total_number_tiles = width * height # total number of tiles
    print ('Total number of tiles is', total_number_tiles)
    
    # compute the number of tiles for each catagory
    tiles_per_category = [round(proportion * total_number_tiles) for proportion in category_proportions]

    # print out number of tiles per category
    for i, tiles in enumerate(tiles_per_category):
        print (central_asia.index.values[i] + ': ' + str(tiles))
    
    # initialize the waffle chart as an empty matrix
    waffle_chart = np.zeros((height, width))

    # define indices to loop through waffle chart
    category_index = 0
    tile_index = 0

    # populate the waffle chart
    for col in range(width):
        for row in range(height):
            tile_index += 1

            # if the number of tiles populated for the current category 
            # is equal to its corresponding allocated tiles...
            if tile_index > sum(tiles_per_category[0:category_index]):
                # ...proceed to the next category
                category_index += 1       
            
            # set the class value to an integer, which increases with class
            waffle_chart[row, col] = category_index
    
    # instantiate a new figure object
    fig = plt.figure()

    # use matshow to display the waffle chart
    colormap = plt.cm.coolwarm
    plt.matshow(waffle_chart, cmap=colormap)
    plt.colorbar()

    # get the axis
    ax = plt.gca()

    # set minor ticks
    ax.set_xticks(np.arange(-.5, (width), 1), minor=True)
    ax.set_yticks(np.arange(-.5, (height), 1), minor=True)
    
    # add dridlines based on minor ticks
    ax.grid(which='minor', color='w', linestyle='-', linewidth=2)

    plt.xticks([])
    plt.yticks([])

    # compute cumulative sum of individual categories to match color schemes between chart and legend
    values_cumsum = np.cumsum(values)
    total_values = values_cumsum[len(values_cumsum) - 1]

    # create legend
    legend_handles = []
    for i, category in enumerate(categories):
        if value_sign == '%':
            label_str = category + ' (' + str(values[i]) + value_sign + ')'
        else:
            label_str = category + ' (' + value_sign + str(values[i]) + ')'
            
        color_val = colormap(float(values_cumsum[i])/total_values)
        legend_handles.append(mpatches.Patch(color=color_val, label=label_str))

    # add legend to chart
    plt.legend(
        handles=legend_handles,
        loc='lower center', 
        ncol=len(categories),
        bbox_to_anchor=(0., -0.2, 0.95, .1)
    )
    plt.show()
```

```python
# Now to create a waffle chart, all we have to do is call the function create_waffle_chart. 
# Let's define the input parameters:

width = 40 # width of chart
height = 10 # height of chart

categories = central_asia.index.values # categories
values = central_asia['Total'] # correponding values of categories

colormap = plt.cm.coolwarm # color map class
```

```python
create_waffle_chart(categories, values, height, width, colormap)
```

# WORD CLOUD
(A depiction of the importance of different words in the body of text)

```python
from wordcloud import WordCloud, STOPWORDS
import urllib
```

```python
# open the file and read it into a variable alice_novel
alice_story = urllib.request.urlopen('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/alice_novel.txt').read().decode("utf-8")
```

```python
stopwords = set(STOPWORDS)
```

```python
# Create a word cloud object and generate a word cloud. For simplicity, 
# let's generate a word cloud using only the first 2000 words in the novel.

# instantiate a word cloud object
alice_word = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate the word cloud
alice_word.generate(alice_story)
```

```python
##Now Visualize it

# display the word cloud
plt.imshow(alice_word, interpolation='bilinear')
plt.axis('off')
plt.show()

# In the first 2000 words in the novel, the most common words are
# Alice, said, little, Queen,etc. 
# Resize the cloud so that we can see the less frequent words better.
```

```python
new_fig = plt.figure(figsize=(12, 14))

# display the cloud
plt.imshow(alice_word, interpolation='bilinear')
plt.axis('off')
plt.show()

# This looks better. But, 'said's is not an informative word. 
# Better add it to stopwords and re-generate the cloud.
```

```python
stopwords.add('said') # add the words said to stopwords

# re-generate the word cloud
alice_word.generate(alice_story)

# display the cloud
new_fig = plt.figure(figsize=(12, 14))

plt.imshow(alice_word, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# save mask to alice_image
alice_image = np.array(Image.open(urllib.request.urlopen('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/labs/Module%204/images/alice_mask.png')))
```

```python
new_fig = plt.figure(figsize=(12, 14))

plt.imshow(alice_image, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# To shape the word cloud according to the mask is simple by using word_cloud package. 
# For ease, lets continue using the first 2000 words in the Alice novel.

# instantiate a word cloud object
alice_word = WordCloud(background_color='white', max_words=2000, mask=alice_image, stopwords=stopwords)

# generate the word cloud
alice_word.generate(alice_story)

# display the word cloud
new_fig = plt.figure(figsize=(14, 18))

plt.imshow(alice_word, interpolation='bilinear')
plt.axis('off')
plt.show()
```

# The immigration data does not have any text data. Let's generate sample text data from our immigration dataset - say text data of 100 words.

```python
canad.head()
```

```python
#Total immigration 1980-2013
total_immig = canad['Total'].sum()
total_immig
```

```python
# Using countries with single-word names, let's duplicate 
# each country's name based on how much they contribute to the total immigration

max_words = 90
word_string = ''
for country in canad.index.values:
    # check if country's name is a single-word name
    if country.count(" ") == 0:
        repeat_num_times = int(canad.loc[country, 'Total'] / total_immig * max_words)
        word_string = word_string + ((country + ' ') * repeat_num_times)

# display the generated text
word_string
```

```python
# create the word cloud
wordcloud = WordCloud(background_color='white').generate(word_string)

print('Word cloud - Country Names')
```

```python
# Show the cloud
plt.figure(figsize=(12, 14))

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

# SEABORN & REGRESSION PLOTS
(Seaborn is useful for drawing statistical graphics such as regression plots, box plots, etc )

```python
import seaborn as sns
```

```python
 # Use the sum() method to get the total population per year
total_immig = pd.DataFrame(canad[years].sum(axis=0))

# change 'years' to type float (Will be convenient for doing regression)
total_immig.index = map(float, total_immig.index)

# reset the index to put in back in as a column in the 'total_immig' dataframe
total_immig.reset_index(inplace=True)

# rename columns
total_immig.columns = ['year', 'total']

# view the final dataframe
total_immig.head()
```

```python
#Display the regression plot
plt.figure(figsize=(14,8))
sns.regplot(x='year', y='total', data=total_immig, color='darkblue', marker='+')
plt.show()
```

```python
##Increase the size of 'markers' and add 'x,y' labels

plt.figure(figsize=(14,8))
ax = sns.regplot(x='year', y='total', data=total_immig, color='darkblue', marker='+', scatter_kws={'s': 100})

ax.set(xlabel='Year', ylabel='Total Immigration') # add x- and y-labels
ax.set_title('Total Immigration to Canada (1993-2013)')

plt.show()                                                                                       
                                                                                              
                                                                                    
```

```python
# Lastly,increase the font size of the tickmark labels, 
# the title, the x- and y-labels so they match the plot size.

plt.figure(figsize=(14,8))

sns.set(font_scale=1.5)

sns.set_style('ticks')  # This changes to white background colour

ax = sns.regplot(x='year', y='total', data=total_immig, color='darkblue', marker='+', scatter_kws={'s': 100})

ax.set(xlabel='Year', ylabel='Total Immigration') # add x- and y-labels
ax.set_title('Total Immigration to Canada (1993-2013)')

plt.show()
```

```python
plt.figure(figsize=(14,8))

sns.set(font_scale=1.5)

sns.set_style('whitegrid')  # This changes to white background colour to gridlines

ax = sns.regplot(x='year', y='total', data=total_immig, color='darkblue', marker='+', scatter_kws={'s': 100})

ax.set(xlabel='Year', ylabel='Total Immigration') # add x- and y-labels
ax.set_title('Total Immigration to Canada (1993-2013)')

plt.show()
```
